```python
# 라이브러리
import time
from Arm_Lib import Arm_Device

# 로봇팔 객체 생성
Arm = Arm_Device()
time.sleep(.1)

# 블록 집기
def arm_clamp_block(enable):
    if enable == 0:
        Arm.Arm_serial_servo_write(6, 60, 400)
    else:
        Arm.Arm_serial_servo_write(6, 130, 400)
        time.sleep(.5)

# 로봇팔의 움직임 제어         
def arm_move(p, s_time = 500): 
    for i in range(5):
        id = i + 1
        if id == 5:
            time.sleep(.1)
            Arm.Arm_serial_servo_write(id, p[i], int(s_time * 1.2))
        elif id == 1:
            Arm.Arm_serial_servo_write(id, p[i], int(3 * s_time / 4))
        else:
            Arm.Arm_serial_servo_write(id, p[i], int(s_time))
        time.sleep(.01)
    time.sleep(s_time / 1000)

# 블록의 초기 위치 및 좌표 설정
p_mould = [90, 130, 0, 0, 90]
p_top = [90, 80, 50, 50, 270]
p_Yellow = [61, 22, 64, 56, 270]
p_Red = [114, 22, 64, 56, 270]
p_Green = [138, 67, 9, 33, 270]
p_Blue = [42, 67, 9, 33, 270]

p_layer_1 = [90, 48, 35, 30, 270]
p_layer_2 = [90, 65, 30, 28, 270]
p_layer_3 = [90, 81, 27, 23, 270]
p_layer_4 = [90, 89, 27, 21, 270]

# 블록 집기 준비
arm_clamp_block(0)
arm_move(p_mould, 1000)
time.sleep(1)

# 노란색 블록 집기
arm_move(p_top, 1000)
arm_move(p_Yellow, 1000)
arm_clamp_block(1)

# 1층에 쌓기
arm_move(p_top, 1000)
arm_move(p_layer_1, 1000) # Brown
arm_clamp_block(0)
time.sleep(.1)
arm_move(p_mould, 1100)

# 빨간색 블록 집기
arm_move(p_top, 1000)
arm_move(p_Red, 1000)
arm_clamp_block(1)

# 2층에 쌓기
arm_move(p_top, 1000)
arm_move(p_layer_2, 1000)
arm_clamp_block(0)
time.sleep(.1)
arm_move(p_mould, 1100)

# 초록색 블록 집기
arm_move(p_top, 1000)
arm_move(p_Green, 1000)
arm_clamp_block(1)

# 3층에 쌓기
arm_move(p_top, 1000)
arm_move(p_layer_3, 1000)
arm_clamp_block(0)
time.sleep(.1)
arm_move(p_mould, 1100)

# 파란색 블록 집기
arm_move(p_top, 1000)
arm_move(p_Blue, 1000)
arm_clamp_block(1)

# 4층에 쌓기
arm_move(p_top, 1000)
arm_move(p_layer_4, 1000)
arm_clamp_block(0)
time.sleep(.1)
arm_move(p_mould, 1100)

# 로봇팔 객체 해제
del Arm
```
