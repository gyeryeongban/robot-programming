```python
# 라이브러리
import time
from Arm_Lib import Arm_Device

# 로봇팔 객체 생성
Arm = Arm_Device()
time.sleep(.1)

# 블록 집기
def arm_clamp_block(enable):
    print(enable)
    if enable == 0:
        Arm.Arm_serial_servo_write(6, 60, 400)
    else:
        Arm.Arm_serial_servo_write(6, 130, 400)
        time.sleep(.5)

# 로봇팔의 움직임 제어
def arm_move(p, s_time = 500):
    for i in range(5):
        id = i + 1
        if id == 5:
            time.sleep(.1)
            Arm.Arm_serial_servo_write(id, p[i], int(s_time * 1.2))
        elif id == 1:
             Arm.Arm_serial_servo_write(id, p[i], int(3 * s_time / 4))
        else:
             Arm.Arm_serial_servo_write(id, p[i], int(s_time))
        time.sleep(.1)
    time.sleep(s_time / 1000)

# 로봇팔의 움직임 제어 수정
def arm_move_modify(p, s_time = 500):
    for i in range(4):
        id = i + 1
        if id == 5:
            time.sleep(.1)
            Arm.Arm_serial_servo_write(id, p[i], int(s_time * 1.2))
        elif id == 1:
             Arm.Arm_serial_servo_write(id, p[i], int(3 * s_time / 4))
        else:
             Arm.Arm_serial_servo_write(id, p[i], int(s_time))
        time.sleep(.1)
    time.sleep(s_time / 1000)

# 로봇팔이 위로 움직임
def arm_move_up():
    Arm.Arm_serial_servo_write(2, 90, 1500)
    Arm.Arm_serial_servo_write(3, 90, 1500)
    Arm.Arm_serial_servo_write(4, 90, 1500)
    time.sleep(.1)

# 블록의 초기 위치 및 좌표 설정
p_mould = [90, 130, 0, 0, 90]
p_top = [90, 80, 50, 50, 270]
p_Brown = [90, 53, 33, 36, 270]
p_Yellow = [65, 22, 64, 56, 270]
p_Red = [115, 22, 64, 56, 270]
p_Red_top = [115, 40, 64, 56, 270]

# 블록 집기 준비
arm_clamp_block(0)
arm_move(p_mould, 1000)
time.sleep(1)

arm_clamp_block(0)
arm_move_up()

# 빨간색 블록 집기
arm_move(p_top, 1000)
arm_move(p_Red, 1000)
arm_clamp_block(1)

# 회색 영역으로 옮기기
arm_move(p_top, 1000)
arm_move(p_Brown, 1000)
arm_clamp_block(0)
arm_move(p_mould, 1100)
time.sleep(.1)

# 빨간색 블록 집기
arm_move_modify(p_Brown, 1000)
arm_clamp_block(1)

# 제자리로 옮기기
arm_move(p_top, 1000)
arm_move_modify(p_Red, 1000)
arm_clamp_block(0)
time.sleep(.2)
arm_move(p_Red_top, 1000)
arm_move(p_mould, 1000)

# 로봇팔 객체 해제
del Arm
```
